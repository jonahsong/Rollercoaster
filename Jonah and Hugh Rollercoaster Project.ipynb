```python
coaster=pd.read_csv('rollercoaster.csv', index_col=0)
```

```python
coaster.columns = coaster.columns.str.strip()
```

```python
bestdata = coaster.drop(columns = ['Drop (feet)', 'Duration (min:sec)', 'G Force', 'Vertical Angle (degrees)', 'CoasterBuzz Ranking'])
bestdata
```

```python
bestcoaster = bestdata.drop(columns = ['Park', 'City/Region', 'City/State/Region', 'Country/Region', 'Geographic Region', 'Inversions (YES or NO)'])
bestcoaster
```

```python
bestcoaster = bestcoaster[bestcoaster.Status == 'Operating']
bestcoaster
```

```python
bestcoaster = bestcoaster.dropna(how = "any")
bestcoaster
```

```python
bestcoaster['Height (feet)'] = bestcoaster['Height (feet)'].astype(float)
heightrating = (bestcoaster['Height (feet)'] - bestcoaster['Height (feet)'].min())/((bestcoaster['Height (feet)'].max() - bestcoaster['Height (feet)'].min())/10)
```

```python
bestcoaster.info()
```

```python
bestcoaster['Hugh and Jonah Index']=4.5*(bestcoaster['Height (feet)']-bestcoaster['Height (feet)'].min())/(bestcoaster['Height (feet)'].max()-bestcoaster['Height (feet)'].min())+5*(bestcoaster['Number of Inversions']-bestcoaster['Number of Inversions'].min())/(bestcoaster['Number of Inversions'].max()-bestcoaster['Number of Inversions'].min())+3.5*(bestcoaster['Speed (mph)']-bestcoaster['Speed (mph)'].min())/(bestcoaster['Speed (mph)'].max()-bestcoaster['Speed (mph)'].min())+2.0*(bestcoaster['Length (feet)']-bestcoaster['Length (feet)'].min())/(bestcoaster['Length (feet)'].max()-bestcoaster['Length (feet)'].min())+1.25*(bestcoaster['Year/Date Opened']-bestcoaster['Year/Date Opened'].min())/(bestcoaster['Year/Date Opened'].max()-bestcoaster['Year/Date Opened'].min())
```

```python
bestcoaster
```

```python
bestcoaster['Speed (mph)'].plot.hist(edgecolor="black", bins=np.arange(bestcoaster['Speed (mph)'].min(),bestcoaster['Speed (mph)'].max(),10))
```

```python
plt.plot(bestcoaster['Length (feet)'],bestcoaster['Number of Inversions'],'.')
plt.xlabel('Length (feet)')
plt.ylabel('Number of Inversions')
```

```python
fit = np.polyfit(bestcoaster['Year/Date Opened'],bestcoaster['Speed (mph)'],1)
fit_fn = np.poly1d(fit)
plt.plot(bestcoaster['Year/Date Opened'],bestcoaster['Speed (mph)'],'.',bestcoaster['Year/Date Opened'], fit_fn(bestcoaster['Year/Date Opened']))
plt.xlabel('Year Opened')
plt.ylabel('Speed (mph)')
```

```python
fit = np.polyfit(bestcoaster['Height (feet)'],bestcoaster['Speed (mph)'],1)
fit_fn = np.poly1d(fit)
plt.plot(bestcoaster['Height (feet)'],bestcoaster['Speed (mph)'],'.',bestcoaster['Height (feet)'],fit_fn(bestcoaster['Height (feet)']))
plt.xlabel('Height (feet)')
plt.ylabel('Speed (mph)')
```

How did you clean the data? What columns/rows did you keep? Which ones did you get rid of? Why? What missing data did you fill in? How? Why?

First, we removed all columns that were missing significant amounts of data. Then we decided that each rollercoaster's rating should not at all relate to the rollercoaster's geographic location because location has nothing to do with the ride itself. So, we removed all of the columns related in anyway to geography. Additionally, we dropped any row (individual coaster) that was still missing any data after the initial removal of various categories. We decided to remove these rows instead of filling them in with a number because there is no way to estimate what this number could be and we thought it would sacrifice the integrity of our ranking system.
